## Creating matplotlib figures using numpy array and more
* Line chart(by default) => plot() , it is by default
* Bar chart => bar()
* Pie chart => pie()
* Histogram => hist()
* Scatter chart => scatter()
* subplots

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
data = np.linspace(0,10,100) #creating 100 data range from 0-10
```

### Line chart


```python
fig, ax = plt.subplots()
plt.plot(data,data**2);
```

### Scatter chart

```python

fig, ax = plt.subplots()
ax.scatter(data,np.exp(data));
```

### Bar chart

```python

fruits_prices = {"apple":200,"grapes":150,"guava":100}
fig,ax = plt.subplots(figsize=(5,3))
ax.bar(fruits_prices.keys(),fruits_prices.values());
ax.set(title="Fruits chart",xlabel="Fruits",ylabel="Prices")
plt.show()
```

### Bar chart with horizontal bars

```python
fruit_prices = {"apple":200,"grapes":150,"guava":100}
fig,ax = plt.subplots()
ax.barh(list(fruit_prices.keys()),list(fruits_prices.values()))
ax.set(title="Fruits Charts")
plt.show()
```

### Pie chart

```python
vehicle_milage = {"swift":32.4,"kwid":28.2,"alto":39.4}
fig,ax = plt.subplots()
ax.pie(vehicle_milage.values(),labels=vehicle_milage.keys(),autopct='%d%%');
ax.set(title="Vehicle Milage's")
plt.show()
```

### Histogram

```python
# labels = ['random'+str(i) for i in range(1,11)]
data = np.random.randint(1,200,10)
fig,ax = plt.subplots()
ax.hist(data)
plt.show()
```

### Subplots
There are two ways to do it.

```python
## 1st way
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(5,5))
ax1.plot(data,data**2);
ax2.scatter(data,np.exp(data))
ax3.bar(['apple','orange','mango'],height=[30,10,50])
ax4.hist(np.random.randint(1,200,10));
```

```python

## 2nd way
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(10,5)) # as ax will be now ndim array
ax[0].plot(data,np.square(data));
ax[1].hist(np.random.randn(1000));
```

## df.plot.hist() vs df.plot()
 df.plot.hist() and df.hist() support specifying the bins parameter directly within the method call. 
 However, df.plot.hist() provides additional flexibility and customization options beyond what df.hist() offers, 
 such as specifying axis labels, title, plot style, and more.

 So, to clarify, both methods allow you to specify bins directly within the method call, but df.plot.hist() offers 
 more extensive customization options beyond just bins

```python
# plotting different all from dataframe in different figure.
heart_disease.plot.hist(figsize=(10,20), subplots=True);
```

## using Object Oriented Method

```python
above_50 = heart_disease[heart_disease['age'] > 50]
above_50
```

```python
len(above_50)
```

```python
# Pyplot method 1
above_50.plot(kind='scatter',x='age',y='chol',c='target') #1st way
```

```python
# Pyplot method 2(aka plotting directly from dataframe plot method)
above_50.plot.scatter(x='age',y='chol',c='target') #2ndway
```

# OO method

```python
above_50.head(1)
```

```python
fig, (ax,ax1) = plt.subplots(figsize=(10,6),nrows=1,ncols=2)
scatter = ax.scatter(above_50['age'],above_50['chol'],c=above_50['target'])
# customization

#set Label
fig.suptitle("Heart Disease Data Set Insight ",fontsize=16,fontweight='bold')
ax.set(title="Age vs Chol",xlabel="Age",ylabel="Chol")
# ax.set_xlabel('Age') # this also works
# ax.set_ylabel('Chol')


# Add legend

# ax.legend(*scatter.legend_elements(), title='Target')
# legend_handles, legend_labels = scatter.legend_elements() #scatter.legend_elements() automatically extracts legend handles and labels based on the colors used in the scatter plot.
# print("=>",legend_handles, legend_labels)

ax.legend(*scatter.legend_elements(), title='Target')

# Add horizontal line 
ax.axhline(above_50['chol'].mean(),ls='--')

# Second Plot
scatter1 = ax1.scatter(above_50['age'],above_50['thalach'],c=above_50['target'])
ax1.set(title="age vs thalach",xlabel="age",ylabel="thalach")
ax1.legend(*scatter.legend_elements(),title='Target')
ax1.axhline(above_50['thalach'].mean())


plt.show()
```
